```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def find_last(self):
        probe = self
        while probe:
            if probe.next is None:
                return probe
            probe = probe.next

    def length(self):
        if self.data is None:
            return 0
        result = 0
        probe = self
        while probe:
            result += 1
            probe = probe.next
        return result


class MyQueue(object):

    def __init__(self):
        self.bucket = Node(None)

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        if self.bucket.data is None:
            self.bucket = Node(x)
        else:
            last_element = self.bucket.find_last()
            last_element.next = Node(x)

    def pop(self):
        """
        :rtype: int
        """
        if self.bucket.data:
            result = self.bucket.data
            self.bucket = self.bucket.next
            if self.bucket is None:
                self.bucket = Node(None)
            return result
        

    def peek(self):
        """
        :rtype: int
        """
        return self.bucket.data

    def empty(self):
        """
        :rtype: bool
        """
        if self.bucket.data is None:
            return True
        return False


# # Your MyQueue object will be instantiated and called as such:
# # obj = MyQueue()
# # obj.push(1)
# # print(obj.bucket.data)
# # param_2 = obj.pop()
# # param_2
# obj.push(3)
# # print(obj.bucket.length())
# # print(obj.bucket.data)
# param_3 = obj.peek()
# param_3
# param_4 = obj.empty()
# param_4
# param_2 = obj.pop()
# print('Length', obj.bucket.length())
# param_2 = obj.pop()
# print('Length', obj.bucket.length())
# print(obj.bucket.data)
# param_4 = obj.empty()
# param_4
```
